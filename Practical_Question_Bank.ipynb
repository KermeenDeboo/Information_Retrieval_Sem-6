**Implement an inverted index construction algorithm for the following two documents**

document1 = "The computer science students are appearing for practical examination."

document2 = "computer science practical examination will start tomorrow."

Build a simple document retrieval system using the constructed index to find the documents containing terms “computer science”.

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Added to download the missing resource
from nltk.tokenize import word_tokenize

docs = {
    "Document 1": "The computer science students are appearing for practical examination.",
    "Document 2": "computer science practical examination will start tomorrow."
}

# 1. Build the Inverted Index
index = {}
for doc, text in docs.items():
    for word in word_tokenize(text.lower()):
        if word.isalpha():
            index.setdefault(word, set()).add(doc)

# 2. Retrieve documents containing "computer" AND "science"
result = index.get("computer", set()) & index.get("science", set())

print(f"Index for 'computer': {index.get('computer')}")
print(f"Index for 'science': {index.get('science')}")
print(f"Retrieved Documents: {list(result)}")
```

**Build a question-answering system using techniques such as information extraction.**

```python
import spacy
from nltk.tokenize import sent_tokenize

# 1. Load the language model and text
nlp = spacy.load("en_core_web_sm")
text = """
Information Retrieval deals with the storage and retrieval of information.
Search engines use IR techniques to retrieve relevant documents.
Natural Language Processing helps computers understand human language.
"""
sentences = sent_tokenize(text)

# 2. Get the question and extract important keywords (ignoring punctuation/stopwords)
question = input("Enter your question: ")
q_keys = {t.text.lower() for t in nlp(question) if not t.is_stop and not t.is_punct}

# 3. Score sentences based on how many keywords they share with the question
scores = {s: sum(1 for t in nlp(s.lower()) if t.text in q_keys) for s in sentences}

# 4. Find the highest scoring sentence and print it (if a match exists)
best_sentence = max(scores, key=scores.get)

print("Answer:", best_sentence if scores[best_sentence] > 0 else "Answer not found.")
```

**Implement a learning to rank algorithm (e.g., RankSVM or RankBoost) on a given dataset.**

```python
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.metrics import ndcg_score

# 1. Define Dataset (Features, Relevance Labels, and Query IDs)
X = np.array([[3, 2, 1], [2, 1, 0], [0, 1, 2], [1, 2, 0], [2, 1, 3], [1, 0, 2]])
relevance = np.array([3, 2, 1, 3, 1, 2])
queries = np.array([1, 1, 1, 2, 2, 2])

# 2. Build Pairwise Training Data (RankSVM Core)
X_pairs, y_pairs = [], []
for q in np.unique(queries):
    idx = np.where(queries == q)[0]
    for i in idx:
        for j in idx:
            if relevance[i] > relevance[j]:
                X_pairs.append(X[i] - X[j]) # Store feature difference
                y_pairs.append(1)           # Label 1: i ranks higher than j
            elif relevance[i] < relevance[j]:
                X_pairs.append(X[i] - X[j])
                y_pairs.append(-1)          # Label -1: i ranks lower than j

# 3. Train the RankSVM Model
model = LinearSVC().fit(X_pairs, y_pairs)

# 4. Evaluate Performance using NDCG
scores = [ndcg_score([relevance[queries==q]], [model.decision_function(X[queries==q])]) for q in np.unique(queries)]
print(f"Average NDCG Score: {np.mean(scores):.4f}")
```

**Train the ranking model using labeled data and evaluate its effectiveness.**

```python
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.metrics import ndcg_score

# 1. Define Labeled Dataset (Features, Relevance, and Query IDs)
X = np.array([[3, 2, 1], [2, 1, 0], [0, 1, 2], [1, 2, 0], [2, 1, 3], [1, 0, 2]])
relevance = np.array([3, 2, 1, 3, 1, 2])
queries = np.array([1, 1, 1, 2, 2, 2])

# 2. Train the Model using Pairwise Differences
X_pairs, y_pairs = [], []
for q in np.unique(queries):
    idx = np.where(queries == q)[0]
    for i in idx:
        for j in idx:
            if relevance[i] > relevance[j]:
                X_pairs.append(X[i] - X[j]) # Feature difference
                y_pairs.append(1)           # Document i is better
            elif relevance[i] < relevance[j]: # Add this condition to create a second class
                X_pairs.append(X[i] - X[j])
                y_pairs.append(-1)          # Document i is worse

# Fit the Linear SVM Classifier
model = LinearSVC().fit(X_pairs, y_pairs)

# 3. Evaluate Effectiveness using NDCG (Normalized Discounted Cumulative Gain)
# Calculate NDCG for each query group
scores = [ndcg_score([relevance[queries==q]], [model.decision_function(X[queries==q])]) for q in np.unique(queries)]

print(f"Average NDCG Score: {np.mean(scores):.4f}")
```

**Implement a text summarization algorithm (e.g., extractive or abstractive).**

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize

# 1. Setup and Download
nltk.download(['punkt', 'stopwords', 'punkt_tab'], quiet=True)
stop_words = set(stopwords.words('english'))

text = """
Information Retrieval is the process of obtaining relevant information from data.
It plays an important role in search engines. Text summarization helps in
reducing document size while preserving key info. Automatic summarization
is widely used in news and research.
"""

# 2. Sentence Tokenization and Word Frequency
sentences = sent_tokenize(text)
words = [w.lower() for w in word_tokenize(text) if w.isalnum() and w.lower() not in stop_words]
word_freq = nltk.FreqDist(words)

# 3. Score Sentences based on word importance
scores = {s: sum(word_freq[w.lower()] for w in word_tokenize(s) if w.lower() in word_freq) for s in sentences}

# 4. Extract Top 2 Sentences
summary = ' '.join(sorted(scores, key=scores.get, reverse=True)[:2])
print("Summary:\n", summary)
```

**Implement the Boolean retrieval model for the following corpus**

Document 1:BSc lectures start at 7.

Document 2:My lectures are over.

Document 3: Today is a holiday. Process the query “not lectures”

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# 1. Define the Corpus
docs = [
    "BSc lectures start at 7.",
    "My lectures are over.",
    "Today is a holiday."
]

# 2. Build the Term-Document Incidence Matrix
# binary=True ensures we only get 1 (present) or 0 (absent)
vectorizer = CountVectorizer(binary=True)
matrix = vectorizer.fit_transform(docs)

# Create a readable DataFrame where columns are words and rows are Doc IDs
df = pd.DataFrame(matrix.toarray(), columns=vectorizer.get_feature_names_out())

# 3. Process the Boolean Query: "NOT lectures"
# Get the 'lectures' column and flip the bits (1 becomes 0, 0 becomes 1)
term = "lectures"
if term in df.columns:
    result_bits = 1 - df[term]  # Bitwise NOT logic
    retrieved_docs = [f"Document {i+1}" for i, val in enumerate(result_bits) if val == 1]

    print("Incidence Matrix for 'lectures':\n", df[term])
    print(f"\nDocuments matching 'NOT {term}':", retrieved_docs)
```

**Implement an inverted index construction algorithm for the following two documents:**

document1 = "best of luck tycs students for your practical examination."

document2 = "tycs students please carry your journal at the time of practical examination."

Build a simple document retrieval system using the constructed index to find the documents containing terms “tycs journal”


```python
# 1. Define the Corpus
docs = {
    "Doc 1": "best of luck tycs students for your practical examination.",
    "Doc 2": "tycs students please carry your journal at the time of practical examination."
}

# 2. Build the Inverted Index
inverted_index = {}
for doc_id, text in docs.items():
    # Tokenize and clean (remove punctuation/lowercase)
    words = text.lower().replace('.', '').split()
    for word in set(words): # Use set to avoid duplicate entries for the same doc
        if word not in inverted_index:
            inverted_index[word] = []
        inverted_index[word].append(doc_id)

# 3. Simple Document Retrieval System
query = "tycs journal"
query_terms = query.lower().split()

# Find documents containing ALL query terms (Intersection)
results = None
for term in query_terms:
    docs_with_term = set(inverted_index.get(term, []))
    if results is None:
        results = docs_with_term
    else:
        results = results.intersection(docs_with_term)

# 4. Display Results
print("Inverted Index Snippet:", {k: inverted_index[k] for k in list(inverted_index)[:5]})
print(f"\nDocuments containing '{query}':", list(results) if results else "No matches found.")
```

**Consider a web graph with the following link structure:**

• Page A has links to pages B and C.

• Page B has a link to page C.

• Page C has links to pages

Apply the PageRank algorithm and analyze the results.


```python
import networkx as nx

# 1. Create a Directed Graph
G = nx.DiGraph()

# 2. Add the link structure (Edges)
# Page A -> B, Page A -> C
# Page B -> C
# Page C has no outgoing links (Dangling Node)
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C')])
G.add_node('C') # Explicitly adding C to ensure it exists

# 3. Apply PageRank Algorithm
# alpha=0.85 is the standard damping factor used by Google
pagerank_scores = nx.pagerank(G, alpha=0.85)

# 4. Analyze and Display Results
print("PageRank Scores:")
for page, score in sorted(pagerank_scores.items(), key=lambda x: x[1], reverse=True):
    print(f"Page {page}: {score:.4f}")
```

• Page A has links to pages B and C.

• Page B has links to pages C and D.

• Page C has links to pages A and D.

• Page D has a link to page B

Apply the PageRank algorithm and analyze the results.

```python
import networkx as nx

# 1. Create a Directed Graph
G = nx.DiGraph()

# 2. Add the link structure (Edges)
# Page A -> B, C
# Page B -> C, D
# Page C -> A, D
# Page D -> B
G.add_edges_from([
    ('A', 'B'), ('A', 'C'),
    ('B', 'C'), ('B', 'D'),
    ('C', 'A'), ('C', 'D'),
    ('D', 'B')
])

# 3. Apply PageRank Algorithm
# alpha=0.85 is the standard damping factor representing the "random surfer"
pagerank_scores = nx.pagerank(G, alpha=0.85)

# 4. Analyze and Display Results
print("PageRank Scores:")
for page, score in sorted(pagerank_scores.items(), key=lambda x: x[1], reverse=True):
    print(f"Page {page}: {score:.4f}")
```

**mplement a text summarization algorithm (e.g., extractive or abstractive).**

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize

# 1. Setup and Download
nltk.download(['punkt', 'stopwords', 'punkt_tab'], quiet=True)
stop_words = set(stopwords.words('english'))

text = """
Information Retrieval is the process of obtaining relevant information from data.
It plays an important role in search engines. Text summarization helps in
reducing document size while preserving key info. Automatic summarization
is widely used in news and research.
"""

# 2. Sentence Tokenization and Word Frequency
sentences = sent_tokenize(text)
words = [w.lower() for w in word_tokenize(text) if w.isalnum() and w.lower() not in stop_words]
word_freq = nltk.FreqDist(words)

# 3. Score Sentences based on word importance
scores = {s: sum(word_freq[w.lower()] for w in word_tokenize(s) if w.lower() in word_freq) for s in sentences}

# 4. Extract Top 2 Sentences
summary = ' '.join(sorted(scores, key=scores.get, reverse=True)[:2])

print("Summary:\n", summary)
```

**Implement a spelling correction module using edit distance for the following dictionary:**
 dictionary = ["spelling", "correction", "algorithm", "information", "retrieval", "system"]

 Integrate the spelling correction module into an information retrieval system.

  Provide suggestions for input word = "information".

```python
import nltk
from nltk.metrics import edit_distance

# 1. Define the Dictionary and Input
dictionary = ["spelling", "correction", "algorithm", "information", "retrieval", "system"]
input_word = "informaton"  # Intentional typo

# 2. Spelling Correction Module
def get_suggestion(word, corpus):
    # Calculate distance for every word in dictionary and find the minimum
    scores = {item: edit_distance(word, item) for item in corpus}
    # Return the word with the lowest edit distance
    return min(scores, key=scores.get)

# 3. Integrate into Retrieval System
suggested_word = get_suggestion(input_word, dictionary)

# 4. Display Results
print(f"Input Word: {input_word}")
print(f"Suggested Correction: {suggested_word}")

if suggested_word in dictionary:
    print(f"\nRetrieving results for: '{suggested_word}'...")
```

**Implement an inverted index construction algorithm for the following two documents:**

document1 = "our class meeting starts soon"

document2 = "my class starts at 6."


```python
# 1. Define the Corpus
docs = {
    "Doc 1": "our class meeting starts soon",
    "Doc 2": "my class starts at 6."
}

# 2. Build the Inverted Index
inverted_index = {}

for doc_id, text in docs.items():
    # Clean and tokenize: lowercase and remove punctuation
    words = text.lower().replace('.', '').split()

    # Use set(words) to avoid duplicate entries for the same document
    for word in set(words):
        if word not in inverted_index:
            inverted_index[word] = []
        inverted_index[word].append(doc_id)

# 3. Display Results
print("Inverted Index Table:")
print("-" * 25)
for term, doc_list in sorted(inverted_index.items()):
    print(f"{term:10} | {', '.join(doc_list)}")
```

**Build a simple document retrieval system using the constructed index to find the documents containing terms “class meeting”.**

```python
# 1. Previously constructed Inverted Index
inverted_index = {
    '6': ['Doc 2'], 'at': ['Doc 2'], 'class': ['Doc 1', 'Doc 2'],
    'meeting': ['Doc 1'], 'my': ['Doc 2'], 'our': ['Doc 1'],
    'soon': ['Doc 1'], 'starts': ['Doc 1', 'Doc 2']
}

# 2. Define the Query
query = "class meeting"
query_terms = query.lower().split()

# 3. Retrieval Logic (Finding common documents)
results = None
for term in query_terms:
    # Get the list of docs for the term, or an empty set if not found
    docs_with_term = set(inverted_index.get(term, []))

    if results is None:
        results = docs_with_term
    else:
        # Keep only documents that appear in all term lists (Boolean AND)
        results = results.intersection(docs_with_term)

# 4. Display Results
print(f"Query: '{query}'")
print(f"Matching Documents: {list(results) if results else 'No matches found.'}")
```

**Create a system that can summarize text data. The given text data is as follows:**

Natural language processing (NLP) is a field of computer science, artificial intelligence, and computational linguistics concerned with the interactions between computers and human (natural) languages. As such, NLP is related to the area of human computer interaction. Many challenges in NLP involve natural language understanding, natural language generation, and machine learning. Text summarization is the process of distilling the most important information from a source (text) to produce an abridged version for a particular user or task. Automatic text summarization methods are greatly needed to address the ever-growing amount of text data available online to both better help discover relevant information and to consume the vast amount of text data available more efficiently.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize

# 1. Setup and Download
nltk.download(['punkt', 'stopwords', 'punkt_tab'], quiet=True)
stop_words = set(stopwords.words('english'))

text = """
Natural language processing (NLP) is a field of computer science, artificial intelligence,
and computational linguistics concerned with the interactions between computers and
human (natural) languages. As such, NLP is related to the area of human computer
interaction. Many challenges in NLP involve natural language understanding, natural
language generation, and machine learning. Text summarization is the process of
distilling the most important information from a source (text) to produce an
abridged version for a particular user or task. Automatic text summarization
methods are greatly needed to address the ever-growing amount of text data
available online to both better help discover relevant information and to
consume the vast amount of text data available more efficiently.
"""

# 2. Tokenize into sentences and words
sentences = sent_tokenize(text)
words = [w.lower() for w in word_tokenize(text) if w.isalnum() and w.lower() not in stop_words]

# 3. Calculate Word Frequencies
word_freq = nltk.FreqDist(words)

# 4. Score Sentences based on word weights
# A sentence's score is the sum of the frequencies of its significant words
scores = {s: sum(word_freq[w.lower()] for w in word_tokenize(s) if w.lower() in word_freq) for s in sentences}

# 5. Extract Top 2 Sentences for the summary
summary = ' '.join(sorted(scores, key=scores.get, reverse=True)[:2])

print("Summary:\n", summary)
```

**Implement the Boolean retrieval model for the following corpus:**

Document 1:The cat chased the dog around the garden.

Document2: She was sitting in the garden last night.

Document 3: I read the book the night before.
            Process the query “garden or night”


```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# 1. Define the Corpus
docs = [
    "The cat chased the dog around the garden.",
    "She was sitting in the garden last night.",
    "I read the book the night before."
]

# 2. Build the Term-Document Incidence Matrix
# binary=True ensures the matrix only contains 1s and 0s
vectorizer = CountVectorizer(binary=True)
matrix = vectorizer.fit_transform(docs)

# Create a DataFrame for clear visualization
df = pd.DataFrame(matrix.toarray(), columns=vectorizer.get_feature_names_out())

# 3. Process the Boolean Query: "garden OR night"
term1, term2 = "garden", "night"

# Check if terms exist in the vocabulary to avoid errors
if term1 in df.columns and term2 in df.columns:
    # Bitwise OR operation: 1 if either term is present
    result_bits = df[term1] | df[term2]

    # Extract Document IDs where the result is 1
    retrieved_docs = [f"Document {i+1}" for i, val in enumerate(result_bits) if val == 1]

    print("Incidence Matrix for query terms:")
    print(df[[term1, term2]])
    print(f"\nDocuments matching '{term1} OR {term2}':", retrieved_docs)
```

**Consider the following corpus**

documents = [ "The cat is on the mat.", "The dog is in the yard.", "A bird is flying in the sky.", "The sun is shining brightly." ]

Build a question-answering system and query for "Where is the cat?"


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Define the Corpus
documents = [
    "The cat is on the mat.",
    "The dog is in the yard.",
    "A bird is flying in the sky.",
    "The sun is shining brightly."
]

query = "Where is the cat?"

# 2. Vectorize using TF-IDF
# This automatically handles tokenization and weights unique keywords like "cat"
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
query_vector = vectorizer.transform([query])

# 3. Calculate Cosine Similarity between the query and each sentence
# similarities[i] = (Query • Document_i) / (||Query|| * ||Document_i||)
similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()

# 4. Extract the most relevant sentence
best_index = similarities.argmax()
answer = documents[best_index]

# 5. Display Results
print(f"Query: {query}")
print(f"Answer: {answer}")
print(f"Confidence (Similarity Score): {similarities[best_index]:.4f}")
```

**Implement an inverted index construction algorithm for the following two documents:**

document1 = "The quick brown fox jumped over the lazy dog"

document2 = "The lazy dog slept in the sun"

Build a simple document retrieval system using the constructed index to find the documents containing terms “lazy sun”.

```python
# 1. Define the Corpus
docs = {
    "Doc 1": "The quick brown fox jumped over the lazy dog",
    "Doc 2": "The lazy dog slept in the sun"
}

# 2. Build the Inverted Index
inverted_index = {}
for doc_id, text in docs.items():
    # Tokenize: lowercase and split
    words = text.lower().split()
    # Use set() to ensure a doc is only listed once per term
    for word in set(words):
        if word not in inverted_index:
            inverted_index[word] = []
        inverted_index[word].append(doc_id)

# 3. Simple Document Retrieval System (Boolean AND)
query = "lazy sun"
query_terms = query.lower().split()

results = None
for term in query_terms:
    docs_with_term = set(inverted_index.get(term, []))
    if results is None:
        results = docs_with_term
    else:
        # Intersect sets to find documents containing ALL query terms
        results = results.intersection(docs_with_term)

# 4. Display Results
print("Inverted Index Snippet:", {k: inverted_index[k] for k in list(inverted_index)[:5]})
print(f"\nDocuments containing '{query}':", list(results) if results else "No matches found.")
```
